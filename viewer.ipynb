```python
import pandas as pd
import matplotlib as plt
```

```python
df2025 = pd.read_csv('df2025.csv')
```

```python
def km(tekst):
    if pd.isna(tekst):
        return None
    try: 
        if "km" in tekst:
            return float(tekst.replace(" km", "").strip())
        if "m" in tekst:
            return float(tekst.replace(" m", "").strip()) / 1000
    except ValueError:
        return None
```

```python
df2025["distance"] = df2025["distance"].apply(km)
```

```python
total2025 = sum(df2025["distance"].dropna())
```
